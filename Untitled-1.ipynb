```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to plot
def func(x):
    return (x - 1) ** 2 - 1

# Generate x values, including integers
x = np.arange(-10, 11)  # Create an array of x values from -10 to 10, inclusive

# Calculate y values based on the function
y = func(x)

# Create a figure and an axes
plt.figure(figsize=(8, 6))

# Plot the function
plt.plot(x, y, label='y = (x - 1)^2 - 1', color='blue')

# Mark the fixed points where x is an integer
for i in range(-10, 11):
    if i.is_integer():
        plt.plot(i, func(i), 'ro', markersize=10, label=f'({i}, {func(i)})')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the function y = (x - 1)^2 - 1 with fixed points')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to plot
def func(x):
    return (x - 1) ** 2 - 1

# Calculate the vertex coordinates
vertex_x = 1
vertex_y = func(vertex_x)

# Generate x values
x = np.linspace(-10, 10, 400)  # Create an array of x values from -10 to 10

# Calculate y values based on the function
y = func(x)

# Create a figure and an axes
plt.figure(figsize=(8, 6))

# Plot the function
plt.plot(x, y, label='y = (x - 1)^2 - 1', color='blue')

# Mark the vertex
plt.plot(vertex_x, vertex_y, 'ro', markersize=10, label=f'Vertex: ({vertex_x}, {vertex_y})')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of the function y = (x - 1)^2 - 1 with vertex')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset

# Set up the main figure
fig, ax = plt.subplots()
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

# Create a circle patch
circle = patches.Circle((0, 0), 2, edgecolor='blue', facecolor='none')
ax.add_patch(circle)

# Set up the inset axes for the cursor coordinates
axins = zoomed_inset_axes(ax, 2, loc=2)  # loc=2 is the same as 'upper right'
axins.set_xlim(ax.get_xlim())
axins.set_ylim(ax.get_ylim())
axins.xaxis.set_visible(False)
axins.yaxis.set_visible(False)

# Function to update the inset axes when the circle is clicked
def update_coordinates(event):
    if circle.contains(event)[0]:
        x, y = event.xdata, event.ydata
        if x is not None and y is not None:
            axins.text(0.5, 0.5, f'({x:.2f}, {y:.2f})', ha='center', va='center', color='red')
    else:
        axins.text(0.5, 0.5, '', ha='center', va='center', color='red')

# Connect the event handler to the 'button_press_event'
fig.canvas.mpl_connect('button_press_event', update_coordinates)

# Display the figure
plt.show()
```

```python
import requests
import json
import hashlib

# 百度翻译API的URL
url = "https://fanyi-api.baidu.com/api/trans/vip/translate"

# 您的API密钥和APP ID
appid = '20190619000308735'
secretKey = '1N_IzdCjfCHrcjiLmIze'

# 函数：根据输入的文本自动选择翻译的方向
def translate_text(text, appid, secretKey, from_lang='auto', to_lang='auto'):
    # 计算签名
    salt = '12345'
    sign = hashlib.md5((appid + text + salt + secretKey).encode()).hexdigest()

    # 发送翻译请求
    params = {
        'appid': appid,
        'q': text,
        'from': from_lang,
        'to': to_lang,
        'salt': salt,
        'sign': sign
    }
    response = requests.get(url, params=params)

    # 解析响应
    trans_result = json.loads(response.text)
    translated_text = trans_result['trans_result'][0]['dst']
    return translated_text

# 用户输入文本
user_input = input("请输入文本：")

# 判断输入的文本第一个字符是英文还是中文
first_char = user_input[0]
if not first_char.isalpha():
    # 如果第一个字符是中文，翻译成英文
    translated_text = translate_text(user_input, appid, secretKey, from_lang='zh', to_lang='en')
else:
    # 如果第一个字符是英文字母，翻译成中文
        translated_text = translate_text(user_input, appid, secretKey, from_lang='en', to_lang='zh')

print(translated_text)
```
